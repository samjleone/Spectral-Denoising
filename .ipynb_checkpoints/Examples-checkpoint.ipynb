# Gaussian Noise

```python
L = G.L.todense()
L_tens = torch.tensor(L)
```

```python
(5*1e-6)*np.trace(L)*np.linalg.norm(noodle_gray)
```

```python
n = G.N
#M = n*torch.eye(n,n) - torch.ones(n,n)
#M = torch.tensor(M, dtype = torch.double)

def loss_fn(z, signal, alpha=1, beta=1, gamma=1):
    smoothness_prior = z.reshape(1,-1)@L_tens@z
    noise_prob = torch.sum(torch.log(torch.abs(z)))
    
    f = signal/x
    KL = uniform_dist(f)
    
    return alpha * noise_prob + alpha * smoothness_prior + gamma * KL
```

```python
def uniform_dist(f):
    K = 500
    h = 5*1e-2
    n = f.shape[0]
    Riemann = torch.tensor(0, dtype = torch.double)

    def kde(z):
        return torch.sum(1/n * 1/torch.sqrt(torch.tensor(2*h**2*torch.pi)) * torch.exp(- (f - z)**2 / (2 * h**2)))

    grid = torch.linspace(-0.5,1.5,K)
    
    for z in grid:
        Riemann += - 1/K * torch.log(kde(z))*kde(z)
        
    estimate = Riemann
    return estimate
```

## Penalty,

$$ \int_0^1 (d(z)-1)^2dz = \int_0^1 (\frac{1}{n}\sum_{x=1}^n k_i(x,z) -1)^2dz $$
$$ = \int_0^1 \bigg[\frac{1}{n}\sum_{x=1}^n \frac{1}{\sqrt{2\pi h^2}}\exp\bigg(-\frac{(x - z)^2}{2h^2}\bigg) -1 \bigg]^2dz $$
$$ = \int_0^1 \bigg[\frac{1}{n}\sum_{x=1}^n \frac{1}{\sqrt{2\pi h^2}}\exp\bigg(-\frac{(x - z)^2}{2h^2}\bigg) -1 \bigg]^2dz $$
$$ = \int_0^1 \bigg[\frac{1}{n}\sum_{x=1}^n \frac{1}{\sqrt{2\pi h^2}}\exp\bigg(-\frac{(x - z)^2}{2h^2}\bigg) -1 \bigg]^2dz $$

## Or KL Divergence: 

$$ \int_0^1 (d(z)-1)^2dz = \int_0^1 u(z) \log(u(z) / d(z)) dz = - \int_0^1 \log(d(z))$$
$$ = - \int_0^1 \log(d(z)) = - \int_0^1 \log(\frac{1}{n}\sum_{x=1}^n \frac{1}{\sqrt{2\pi h^2}}\exp\bigg(-\frac{(x - z)^2}{2h^2}\bigg) $$
$$ = - \int_0^1 \log(d(z)) = - \log(\frac{1}{n\sqrt{2\pi h^2}}) - \int_0^1 \log\bigg(\sum_{x=1}^n \exp\bigg(-\frac{(x - z)^2}{2h^2}\bigg)\bigg) $$

Which can be estimated via: 

$$ \log\bigg(n\sqrt{ 2\pi h^2}\bigg) + \frac{1}{K} \sum_{z=0}^{K-1} \log\bigg(\sum_{i=1}^n \exp\bigg(-\frac{(x(i) - z/K)^2}{2h^2}\bigg)\bigg) $$ 

```python
original = noodle_gray.reshape(-1)
original = original/np.linalg.norm(original)
factors = np.random.uniform(0,1,G.N)
noisy = original * factors
x = torch.tensor(noisy, requires_grad=True)
signal = torch.tensor(noisy)
```

```python
opt = optim.Adam([x], lr = 1e-2)
```

```python
for i in range(200):
    loss = loss_fn(x, signal,alpha=5*1e-6, beta=1, gamma=5*1e-2)
    loss.backward()
    opt.step()
    norm = torch.norm(x.grad)
    opt.zero_grad()
    
    with torch.no_grad():
        for j in range(x.shape[0]):
            if x[j]/noisy[j] < 1:
                x[j] = noisy[j]
    
    if i % 25 == 0:
        #print(norm)
        #print(i)
        print("Optimizing", loss)
        print("Signal Correlation:", np.corrcoef(original,x.detach().numpy().reshape(-1))[0,1])
```

```python
z= torch.tensor(original/np.linalg.norm(original))
loss_fn(z, signal, alpha = 5*1e-6)
```
